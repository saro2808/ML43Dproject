```python
# For automatic file reloading as usual
%load_ext autoreload
%autoreload 2
%matplotlib inline

from pathlib import Path
import torch

from utils import *
from renderer import *

device = torch.device("cuda")

root_path = Path("data/replica")
all_scenes = ["office0","office1","office2","office3","room0",
              "office4","room1","room2"]
train_scenes = all_scenes[:6]
eval_scenes = all_scenes[6:]
```

```python
K = load_intrinsics(root_path / all_scenes[0] / "intrinsics.txt")
# array([[317.5,   0.        , 319.5 ],
#        [  0. , 317.64705882, 179.5 ],
#        [  0. ,   0.        ,   1.  ]])
print(K)
poses = load_poses(root_path / all_scenes[0] / "poses")
assert len(poses) == 200
images = load_images(root_path / all_scenes[0] / "color")
H, W, C = images[0].shape  # (360, 640, 3)
assert len(images) == 200
cameras = [
    build_camera_from_replica(K, pose, H, W) for pose in poses
]
```

```python
mesh_path = "data/replica/office0/office0_mesh.ply"

mesh = load_mesh(mesh_path, device)
```

## RGB rendering

```python
rgb_renderer = RgbRenderer(H, W, device)
```

```python
rgb_images = []
for cam in cameras[:5]:
    try:
        rendered_images = rgb_renderer(mesh, cameras=cam)
    except Exception:
        continue
    rgb_images.append(rendered_images[0, ..., :3].cpu().numpy())  # HxWx3

len(rgb_images)
```

```python
mask_renderer = MaskRenderer(H, W, device)

# render masks
mask_images = []
for cam in cameras[:5]:
    masks = mask_renderer(mesh, cameras=cam)
    mask_images.append(masks[0, ..., 3].cpu().numpy())  # alpha channel
```
